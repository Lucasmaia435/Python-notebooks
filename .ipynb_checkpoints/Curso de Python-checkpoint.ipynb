### Listas

Listas são aglomerados de dados em uma dimensão, ou seja, um grupo de diferentes dados, em um espaço, e nesse espaço podemos ter números, strings ou até mesmo outras listas, podendo ter o valor de seus elementos alterados durante a execução do programa. Listas sempre são declaradas entre colchetes.

```python
lista1 = [1,2,3,4,5] # Aqui temos uma lista que tem 5 elementos numéricos
```

```python
lista2 = [10, "Segundo elemento",[11, "Segundo"]] # Aqui temos uma lista com elementos numéricos, strings e outra lista
```

Para podermos escolher elementos específicos da lista, podemos procurar pelo seu indice. OBS.: O indice sempre começa de 0, ou seja o primeiro elemento da lista, é o elemento de índice 0.

Observe o exemplo:

```python
lista1[0] # Buscando o elemento de índice 0, ou seja, o primeiro elemento da lista
```

```python
lista1[0] = 10 # Alterando o valor do primeiro elemento da lista para 10
```

```python
lista1 # Mostrando a alteração no valor no elemento lista1[0]
```

```python
lista2[2] # Nesse caso, o elemento de índice 2 da lista2 é outra lista, e se quisermos pegar um elemento dessa segunda
          # lista? Observe o próximo bloco de código
```

```python
lista2[2][1] #Essa linha de código pegamos o terceiro elemento(índice 2) da 'lista2' e depois pegamos o segundo elemento
             #da lista, que é o terceiro elemento da 'lista2'
```

* ### Revisando operações lógicas

 Se você nunca estudou sobre operações lógicas essa parte vai ser ótima para você, já que não dependemos delas apenas para armazenar valores True ou False em uma variável, mas também para diversas situações na área da computação. Então vamos rapidamente entender as operações And e Or. 

* A operação And, traduzida do inglês "e", é usada para compararmos N situações, em que todas devem ter o mesmo valor lógico, ou seja, se

                               (num1 > num2) = True   e    (num3 > num4) = True
                                                    Então
                                    ((num1 > num2) and (num3 > num4)) = True
                                      

* A operação Or, traduzida do inglês "ou", é usada para compararmos N situações, em que apenas uma deve ter o valor lógico True para ser verdadeiro.

                               (num1 < num2) = False   e    (num3 > num4) = True
                                                    Então
                                    ((num1 < num2) or (num3 > num4)) = True
                                      

A imagem a seguir, representa uma tabela verdade com as operaçoes And e Or. Considere T = True e F = False.

E então concluímos que na operação And, todas as situações devem ser iguais para termos um resultado True. Já na operação or, precisamos apenas de uma única operação verdadeira, para obtermos um resultado True.

## Estruturas condicionais 

Já que agora nós sabemos como fazer a comparação entre duas variáveis, elementos ou situações, então vamos aprender agora 3 estruturas que chamamos de condicionais, que são:
* If
* Else
* Elif

### If


O *If* é uma estrutura de verificação de condição, que enquanto a condição que lhe foi concebida, for verdadeira, ela irá executar determinadas instruções.

```python
num1 = int(input()) # Declamaramos uma variável do tipo inteiro recebida pelo usuário
```

```python
if num1 > 9:  # Ao lado do if, colocamos a nossa condição, que nesse caso é "num1 é maior que 9?"
    print(num1) # Se a condição retornar verdadeiro, então ele entra na estrutura que é representada pela indentação
    print("A condição é verdadeira")
    #até haver uma quebra na indentação do if, tudo irá estar dentro do if 
    
    
    
    # Dentro do if
#Fora do if
```

### Else

O *Else* é uma estrutura auxiliar do *If*, as traduções de "If" e "Else", são "Se" e "Senão". Então o *Else* é o que ele deve fazer, se a condição do *If* retornar falso.

```python
num2 = int(input()) # Declamaramos uma variável do tipo inteiro recebida pelo usuário
```

```python
if num2 >= 11: # Verifica se num2 é maior que 11
    print(num2) # Se a condição do If fosse saciada, então iriamos printar o num2
    
    # Dentro do if
    
else: # Aqui, nós saimos do If e fomos para o Else, já que a condição não foi saciada
    print("11 é maior") # Printando "11 é maior" já que o num2 é menor que 11
    
    #Dentro do else
    
#Fora do else
```

### Elif

O *Elif* é um *If* misturado com o *Else*, por isso "elif". O *Elif* é usado normalmente quando você tem diversas condições, além de uma condição e sua contra-condição.

```python
num3 = int(input()) # Declamaramos uma variável do tipo inteiro recebida pelo usuário
```

```python
if num3 > 11: # Verificamos se 'num3' é maior que 11 / Primeira condição
    print("Maior") # Se a primeira condição for saciada, esse comando será executado

    # Dentro da primeira condição
elif num3 == 11: # Verificamos se 'num3' é igual à 11 / Segunda condição
    print("Igual")# Se a segunda condição for saciada, esse comando será executado

    # Dentro da segunda condição
else:
    print("Menor") # Se nenhuma das condições for saciadas, então esse comando será executado
    
    # Dentro do else
```

## Estruturas de repetição

As estruturas de repetição são usadas para que não tenhamos que repetir diversas vezes a mesma operação, por exemplo, printar 100 vezes um nome, ao invés de escrevermos 100 linhas com "print('nome')", iremos usar essas novas estruturas
para repetir a mesma instrução quantas vezes forem necessárias. 

As duas estruturas de repetição que iremos estudar são:
* While
* For

### While

O *While* é uma estrutura de repetição que se assemelha com o if, devido a presença de uma condição, então, enquanto a condição for verdadeira, as instruções que estão dentro do *While* irão se repetir até que a condição seja falsa.

**Contadores** são variáveis que normalmente são usadas apenas como parâmetros, então, a unica operação que nós iremos realizar nele será a soma de +1, para podermos saber a quantidade de vezes que determinada instrução ou estrutura foi executada. E podemos usar contadores para definirmos um *While* 'contado', que seria um *While* que acontece até que nosso contador alcance certo valor.


```python
i = 1 # Iniciamos nosso contador i, que é apenas uma variável usada como referência.
while i < 6: # Então, aqui é a nossa condição, que é lida da seguinte forma "Enquanto i for menor que 5 :"
    # Começo do while, tudo a baixo daqui, será repetido até que a condição seja falsa.
    
    print(i) # Enquanto i(contador) for menor que 5, ele irá ser printado.
    i = i + 1 # O contador tendo seu valor somado à 1, realizando assim a "contagem" desejada
    
    # Assim como a estrutura condicional todas as estruturas, sejam elas de repetição ou funções, que veremos mais adiante
    # Funcionam com a indentação
    
    # Se a condição no final do while, ainda for verdadeira, o programa irá retornar para o começo do while
    # gerando assim, a repetição
    
print("Fora da repetição")
```

Outro uso para o while, sem o uso do contador, é realizar uma instrução até que a condição for falsa

```python
num = int(input()) # Solicitamos ao usuário um número inteiro
while num != 10: # Então verificamos se o número digitado pelo usuário é igual à 10, se ele não for então a estrutura será 
                 # executada até que o número digitado seja 10.
    
    print('Digite o número 10') # Informa ao usuário que o programa quer receber o número 10
    num = int(input()) # Pede ao usuário novamente um número, e se não for igual à 10, a estrutura irá se repetir
    

print('Pronto, o número digitado é igual à 10') # Quando o usuário passar um número igual a 10, então o while irá parar 
                                                # e então irá printar essa ultima linha
```

### For

O *For* realiza a repetição da mesma maneira do *While*, porém, não é uma condição que causa a repetição, e sim o fim do *For*, porque para que nós possamos utilizar o *For* é necessário que tenha um inicio e um fim, como um intervalo matemático, ou seja, o programa sabe quando vai parar. 

#### Definindo um inicio e um fim

Juntamente ao *For*, iremos também aprender a função range(), que serve para criar um intervalo, ou seja, o inicio e o fim que precisamos para o for. Esse inicio e fim também podem ser determinados por uma lista, mas iremos focar aqui usando o range().

a função range() tem 1,2 ou 3 argumentos, dependendo da sua necessidade. 

Quando usamos apenas um argumento o range vai de 0 até o argumento-1 :
        

```python
range(5) # de 0 até 4 indo em 1 em 1 
```

Se usarmos 2 argumentos, o range vai do primeiro argumento até o segundo argumento - 1:

```python
range(1,4) # de 1 até 3 indo em 1 em 1
```

Se usarmos 3 argumentos, o range vai do primeiro argumento até o segundo-1, usando o 3 como incrementador. Quando não definimos o incrementador o range vai de 1 em 1, do inicio ao fim, se definirmos o incrementador como x, por exemplo, ele vai do inicio ao fim, contando de x em x :

```python
range(0,5,2) # De 0 à 4 indo de 2 em 2
```

#### Voltando ao For

Vamos experimentar os nosso range's que usamos agora a pouco como exemplos

```python
for i in range(5): # Lemos esse trecho da seguinte forma 'Para i em 0 até (5-1), com i = i + 1'
    print(i)
```

```python
for i in range(1,4):# Lemos esse trecho da seguinte forma 'Para i em 1 até (4-1), com i = i + 1'
    print(i)
```

```python
for i in range(0,5,2):# Lemos esse trecho da seguinte forma 'Para i em 0 até (5-1), com i = i + 2'
    print(i)
```

Agora aplicando o *For* em uma situação real
 - Uma função exponencial matematica é dada por f(x) = (a^x), faça um programa que receba um valor para a e dê valores para f(x), com x variando de 0 até 5

```python
a = float(input())
print('--------')
for x in range(5):
    print(('f({func}): {result}').format(func=x,result=a**x)) # Sempre 'a' será igual, porém x vai variar de 0 até 4
```

## Funções

Funções são estruturas que realizam determinadas instruções pelo chamado de apenas uma instrução que seria o nome da função, normalmente é usado quando utilizam a mesma instrução ou conjunto de instruções em diferentes partes do código, sem necessitar a sua descrição

```python
var = 10 # Variavel que será usada como argumento

def mostre(arg1): # Declaração da função junto com o argumento que irá ser usado
    return print(arg1) # O que a função irá retornar, pode ser um valor, variavel ou comando
    # Dentro da função

# Fora da função

mostre(var) # Chamamos a função, usando a variável 'arg' como argumento para ela
```

```python
num1 = 10 # Variaveis com valores arbitrários que serão usados 
num2 = 5

def soma(arg1,arg2): # Função que irá receber o valor de 2 argumentos
    
    soma = arg1+arg2 # Variavel local da função que recebe a soma dos 2 argumentos recebidos
    
    return soma # A função retorna o valor da variavel 

num3 = soma(num1,num2) # A variável 'num3' irá receber o valor retornado da função 'soma' quando os argumentos são as
                       # variáveis 'num1' e 'num2'
print(num3)
```

### Aplicando funções

```python
print("Bem vindo à calculadora de função de primeiro grau!")
# Agora iremos escrever a função que irá realizar nosso calculo
def func1grau(a,b,xi,xf): # Declaramos uma função de uma equação de primeiro grau, que irá receber como argumentos 
                          # os valores de a,b e o intervalo de X.
    print("Os valores da função em função de 'x':")
    for i in range(xi,xf+1):
        func = a*i + b
        print("f({cont}) = {resultado}".format(cont=i,resultado=func))

# Aqui já estamos fora do corpo da função, então podemos voltar ao código principal
print("Digite os valores de a e b para a função:")
a = float(input()) # Recebmos as variáveis que depois iremos usar como argumentos para a função
b = float(input())
print("Agora digite o intervalo de valores para x, ou seja, do primeiro valor, até o segundo")
inicio = int(input())
fim = int(input())

func1grau(a,b,inicio,fim) # Aqui, chamamos a função, com os argumentos que queremos. Lembre que os argumentos devem estar
                          # igualmente posicionados aos que colocamos na definição da função
```
